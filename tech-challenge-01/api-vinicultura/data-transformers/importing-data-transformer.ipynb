```python
import pandas as pd
years = [str(year) for year in range(1970, 2023)]
columns = [*["id", "País"], *years]

df = pd.read_csv("../datasource/csv/ImpVinhos.csv", sep=";", encoding="utf-8")
year_columns = [col for col in df.columns if col.isdigit()]
print(df.columns)

df.head()
```

```python
df_melted = df.melt(id_vars=["Id", 'País'], var_name='ano', value_name='valor')
df_grouped = df_melted.groupby('ano').apply(lambda x: x.groupby(x.columns, axis=1).sum()).reset_index(drop=True)

print(df_grouped)

df_grouped.to_csv("Import_test_transformed.csv", index=False)
```

```python
import uuid

df_to_transform = df_grouped.copy()

categoria_atual = None
for idx, row in df_to_transform.iterrows():
    if row['control'].isupper():
        categoria_atual = row['produto']
    else:
        df_to_transform.loc[idx, 'categoria'] = categoria_atual

df_final = df_to_transform[df_to_transform['categoria'].notnull()]
df_final['product_id'] = [uuid.uuid4() for _ in range(len(df_final))]

# df_final.to_csv("datasource/csv/Comercio_transformed.csv", index=False)

df_final.head()
```
