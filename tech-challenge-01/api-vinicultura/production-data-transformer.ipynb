```python
import pandas as pd
df = pd.read_csv("datasource/csv/Producao.csv", delimiter=";")
df.head()
```

```python
df.head()
```

```python
df_melted = df.melt(id_vars=['id', 'produto'], var_name='ano', value_name='valor')

print(df_melted)

indice_vinho_de_mesa = df[df['produto'] == 'VINHO DE MESA'].index[0]
indice_vinho_fino = df[df['produto'] == 'VINHO FINO DE MESA (VINÍFERA)'].index[0]
print("indice_vinho_de_mesa", indice_vinho_de_mesa, "indice_vinho_fino", indice_vinho_fino)

# # Dividir os dados em duas tabelas
# df_produto = df_melted[['id', 'produto']].drop_duplicates().reset_index(drop=True)
# df_vendas = df_melted[['id', 'ano', 'valor']]
```

```python
import uuid

df_to_transform = df_melted.copy()

categoria_atual = None
for idx, row in df_to_transform.iterrows():
    if row['produto'].isupper():
        categoria_atual = row['produto']
    else:
        df_to_transform.loc[idx, 'categoria'] = categoria_atual

df_final = df_to_transform[df_to_transform['categoria'].notnull()]
df_final['product_id'] = [uuid.uuid4() for _ in range(len(df_final))]

df_final.to_csv("datasource/csv/Producao_transformed.csv", index=False)

df_final.head()
```

```python
import sqlite3

conn = sqlite3.connect('dados.db')

# Criação de um cursor
cur = conn.cursor()

# Criação da tabela de produtos e vendas
cur.execute('''CREATE TABLE IF NOT EXISTS produtos_vendas
            (product_id TEXT PRIMARY KEY, produto TEXT, categoria TEXT, ano INTEGER, valor INTEGER)''')

# Inserção dos dados na tabela de produtos e vendas
for idx, row in df_final.iterrows():
    cur.execute(
        "INSERT INTO produtos_vendas (product_id, produto, categoria, ano, valor) VALUES (?, ?, ?, ?, ?)",
        (str(row['product_id']), row['produto'], row['categoria'], row['ano'], row['valor'])
    )

# Commit das transações
conn.commit()

# Fechamento do cursor e da conexão
cur.close()
conn.close()
```

```python
import sqlite3
from tabulate import tabulate

# Conexão com o banco de dados SQLite
conn = sqlite3.connect('dados.db')

# Criação de um cursor
cur = conn.cursor()

# Execução da query
cur.execute('SELECT produto, SUM(valor), categoria FROM produtos_vendas GROUP BY categoria;')

# Obtenção dos resultados
rows = cur.fetchall()

# Fechamento do cursor e da conexão
cur.close()
conn.close()

# Imprimir os resultados em formato de tabela
headers = [desc[0] for desc in cur.description]
print(tabulate(rows, headers=headers, tablefmt='grid'))
```
